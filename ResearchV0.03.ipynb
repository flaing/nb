# 具體建議 #

## 基本功能 ##

> 搜索的目的就是節省顧客找到他真正要的東西的時間，如果能在短時間達到比價, 目標商品而趨近真實需要，顧客就願意付錢。但顧客往往犯拼寫錯誤或丟一個模糊的搜索條件，好的搜索服務就是讓顧客即使犯錯也能有好的結果。

目前台蛋直接使用資料庫全文檢索，效率慢、消耗系統資源，功能上也難達成客戶需要。因此應該貫徹「資料庫建索引，索引丟搜索」的設計原則。

##一些驗收時的重要指標：

 * search is social, sensor is creator of contents, too.
* Speed # just speed...
* Relevance tuning # ranking, scoring
* Navigation and filtering # sort and limit
* Federated Search # index things
* Linguistic toolset # vocabularies -(nlp)-> autocategorization
* Search analytics # user behavior

##框架選用：

至於選擇框架上，我們過去有 solr 的使用經驗，但考量到 Multiple document types per schema 和 Online schema changes，建議選用 schema-free 的 elasticsearch，因為台蛋網尚在摸索階段很可能會在資料庫改動許多，這樣的彈性能為我們省下邊際的成本; Solr 應該在我們的實驗時期結束再引入。

##關聯打分：

由於目前台蛋網商品規模不大，很難建立很好的關聯性算法，而給予客戶優質的搜索結果，因此建議在算法中加入 entropy 或 evedence 等參數，初期可以由 marketing 或相關業務團隊觀察正確性之餘，手動鍵入修正。

##雙排序：

使用者追求最便宜、價錢又最低的，但很少搜索支援這種巢狀功能的搜索，就是譬如說，我先點'價格:低->高'再點'關聯性:高->低'的這樣兩層的巢狀關係。藉此，我能快速的找到我要的商品。為了支援這樣的功能，減少伺服器負載，提升體驗，所以也建議"搜索" query 的時候，就先撈回 rank 較高在最前面的前 50- 100 項回客戶端，除非後續客戶端搜索的條件有巨大改變，才再與伺服器溝通一次。

## 意見:
如果說智能搜索特點分為{智能化, 個性化, 移動化, 主動性, 交互性}
我認為其中重要的是1.[移動化、交互性]，次之是2.[智能化]，再來是3.[個性化、主動性]。原因是會在店商買東西的人，大多心理有譜了，個性化並不是首要的問題，卻會耗費我們大量研發成本，然後客人大多遍歷不同網站比價，並且會切換不同裝置，所以針對裝置log優化的部分也還好。但目前我們的移動性和交互性則是迫切需要改善的問題。
移動性部分，雖然我們有手機 client ，但對大多數用戶來講查找產品一定是透過搜尋入口網站大量比價的，會下載來用的估計也是熟客了，那麼為什麼不要針對 mob.web 的需求做 RWD 呢? 這樣既不會影響 SEO ，又能讓客戶用的舒服。台灣的服務譬如 ASAP 沒做這塊，我們要是做了也會比較有優勢，現在大多數人的消費瀏覽行為都是用 mob.web 了，移動性我認為相當重要，再來就是針對 mob.web 設定一些行銷目的或遊戲，帶給客戶一些不同的體驗，因為電商的行銷與搜索應當是綁在一起密不可分的，但這是題外話了。
