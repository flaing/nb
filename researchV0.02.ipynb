## 意見:
如果說智能搜索特點分為{智能化, 個性化, 移動化, 主動性, 交互性}
我認為其中重要的是1.[移動化、交互性]，次之是2.[智能化]，再來是3.[個性化、主動性]。原因是會在店商買東西的人，大多心理有譜了，個性化並不是首要的問題，卻會耗費我們大量研發成本，然後客人大多遍歷不同網站比價，並且會切換不同裝置，所以針對裝置log優化的部分也還好。但目前我們的移動性和交互性則是迫切需要改善的問題。
移動性部分，雖然我們有手機 client ，但對大多數用戶來講查找產品一定是透過搜尋入口網站大量比價的，會下載來用的估計也是熟客了，那麼為什麼不要針對 mob.web 的需求做 RWD 呢? 這樣既不會影響 SEO ，又能讓客戶用的舒服。台灣的服務譬如 ASAP 沒做這塊，我們要是做了也會比較有優勢，現在大多數人的消費瀏覽行為都是用 mob.web 了，移動性我認為相當重要，再來就是針對 mob.web 設定一些行銷目的或遊戲，帶給客戶一些不同的體驗，因為電商的行銷與搜索應當是綁在一起密不可分的，但這是題外話了。
