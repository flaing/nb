```python
import solr

# create connection
s = solr.SolrConnection('http://10.16.197.77:8083/solr')

# add a document to the index
s.add(id=1, title='eat newegg', creator=['old Chicken', 'old duck'])
s.commit()
```

```python
# search
response = s.query('title:lucene')
for hit in response.results:
    print hit['title']
```
